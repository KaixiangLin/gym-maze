```python
%matplotlib inline
from IPython.display import display, HTML

from Gym_GridWorld.envs import GridWorldEnv
from Gym_GridWorld.envs import RandomGridGenerator, TMazeGenerator, WaterMazeGenerator
```

# Simple empty grid world

```python
randomgrid = RandomGridGenerator(grid_size=4, obstacle_ratio=0.0)
env = GridWorldEnv(randomgrid, trace=True)
env._reset(solve=True)

for action in env.optimal_solution[('optimal actions', 0)]:
    env.step(action)
    fig = env.render()
    
HTML(env._get_video(interval=200, gif_path='./tmp.gif').to_html5_video())
```
