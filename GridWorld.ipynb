```python
%matplotlib inline
from IPython.display import display, HTML

from Gym_GridWorld.envs.utils import RandomGridGenerator, TMazeGenerator
from Gym_GridWorld.envs.gridworld import GridWorldEnv
```

```python
randomgrid = RandomGridGenerator(grid_size=30, obstacle_ratio=0.2)
sampled_state = randomgrid.sample_state()
gridworld = GridWorldEnv(randomgrid.grid, init_state=sampled_state[0], goal_states=sampled_state[1])
gridworld._reset()

for action in gridworld.optimal_solution[('optimal actions', 0)]:
    gridworld._step(action)
    fig = gridworld._render()
    
HTML(gridworld._get_video(interval=200, gif_path='./tmp.gif').to_html5_video())
```

```python
T_maze = TMazeGenerator(5, [5, 3], [3, 3])
init_state, goal_states = T_maze._sample_state()
gridworld = GridWorldEnv(grid_map=T_maze.maze, init_state=init_state, goal_states=goal_states)
gridworld._reset()

for action in gridworld.optimal_solution[('optimal actions', 0)]:
    gridworld._step(action)
    fig = gridworld._render()
    
HTML(gridworld._get_video(interval=200, gif_path='./tmp.gif').to_html5_video())
```
