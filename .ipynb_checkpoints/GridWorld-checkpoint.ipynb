```python
%matplotlib inline
from IPython.display import display, HTML

from Gym_GridWorld.envs import GridWorldEnv
from Gym_GridWorld.envs import RandomGridGenerator, TMazeGenerator
```

# Random grid world

```python
randomgrid = RandomGridGenerator(grid_size=30, obstacle_ratio=0.2)
gridworld = GridWorldEnv(randomgrid)
gridworld.reset()

for action in gridworld.optimal_solution[('optimal actions', 0)]:
    gridworld.step(action)
    fig = gridworld.render()
    
HTML(gridworld._get_video(interval=200, gif_path='./tmp.gif').to_html5_video())
```

# Multiple T-maze

```python
T_maze = TMazeGenerator(3, [5, 3], [3, 3])
gridworld = GridWorldEnv(T_maze)
gridworld.reset()

for action in gridworld.optimal_solution[('optimal actions', 0)]:
    gridworld.step(action)
    fig = gridworld.render()
    
HTML(gridworld._get_video(interval=200, gif_path='./tmp.gif').to_html5_video())
```
